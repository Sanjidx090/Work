```python
# prompt: add a new column 'post no' based on the text,,each text starts with #2443,,like this

import re

# Function to extract the post number from the text
def extract_post_number(text):
  if isinstance(text, str):
    match = re.search(r'#(\d+)', text)
    if match:
      return int(match.group(1))
  return None # Or handle cases where the pattern is not found

# Apply the function to the 'text' column to create the 'post no' column
df['post no'] = df['text'].apply(extract_post_number)

# Display the first few rows with the new column
display(df.head())
print("\nDataFrame with 'post no' column:")
print(df.columns)
print("\nFirst few values of 'post no':")
print(df['post no'].head())
```

```python
# prompt: texts are like  '#2443
# I wanna write “I miss you” on a rock and throw it to your face, just so you know how much it hurts to miss you : )
# ~an angry bird'..........get some info here from here,,like clean  the number ,,remove the \n and place line brake or space there,,,you can crate a sender column ,for instace 'an angry bird' here

def clean_text_and_extract_sender(text):
  """
    Cleans the text by removing the leading post number and newline characters,
    and extracts the sender from the last line if present.

    Args:
      text: The input text string.

    Returns:
      A tuple containing:
        - The cleaned text string.
        - The sender string, or None if no sender is found.
    """
  if not isinstance(text, str):
    return text, None # Return original value and None if not a string

  # Remove the leading post number like '#2443'
  cleaned_text = re.sub(r'^#\d+\s*', '', text)

  # Split the text into lines
  lines = cleaned_text.split('\n')

  # Check the last line for a potential sender (starts with ~)
  sender = None
  if len(lines) > 1 and lines[-1].strip().startswith('~'):
    sender = lines[-1].strip()[1:].strip() # Remove ~ and leading/trailing whitespace
    cleaned_text = '\n'.join(lines[:-1]) # Join all lines except the last one

  # Replace remaining newline characters with spaces
  cleaned_text = cleaned_text.replace('\n', ' ')

  return cleaned_text, sender

# Apply the function to the 'text' column
df[['cleaned_text', 'sender']] = df['text'].apply(lambda x: pd.Series(clean_text_and_extract_sender(x)))

# Display the first few rows with the new columns
display(df.head())
print("\nDataFrame with 'cleaned_text' and 'sender' columns:")
print(df.columns)
print("\nFirst few values of 'cleaned_text':")
print(df['cleaned_text'].head())
print("\nFirst few values of 'sender':")
print(df['sender'].head())
```
