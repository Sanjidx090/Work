```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
import torch

# Load dataset
df = pd.read_csv("/content/train (1).csv")

# Preprocess text data
df['text'] = df['text'].str.lower()

# Encode sentiment labels
label_encoder = LabelEncoder()
df['sentiment'] = label_encoder.fit_transform(df['sentiment'])

# Split dataset into training and validation sets
train_texts, val_texts, train_labels, val_labels = train_test_split(df['text'], df['sentiment'], test_size=0.2)

# Load pre-trained BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')

# Tokenize text data
train_encodings = tokenizer(train_texts.tolist(), truncation=True, padding=True)
val_encodings = tokenizer(val_texts.tolist(), truncation=True, padding=True)

# Convert data to PyTorch format
class SentimentDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = SentimentDataset(train_encodings, train_labels)
val_dataset = SentimentDataset(val_encodings, val_labels)

# Load pre-trained BERT model
model = BertForSequenceClassification.from_pretrained('bert-base-multilingual-cased', num_labels=len(label_encoder.classes_))

# Define training arguments
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
)

# Define trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
)

# Train model
trainer.train()

# Save the trained model
model.save_pretrained('./sentiment_model')
tokenizer.save_pretrained('./sentiment_model')
```

```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification
import torch

# Load the modified dataset
df = pd.read_csv("/content/test - train.csv")

# Load the saved model and tokenizer
model = BertForSequenceClassification.from_pretrained('./sentiment_model')
tokenizer = BertTokenizer.from_pretrained('./sentiment_model')

# Tokenize text data
encodings = tokenizer(df['text'].tolist(), truncation=True, padding=True, return_tensors='pt')

# Predict sentiments
model.eval()
with torch.no_grad():
    outputs = model(**encodings)
    predictions = torch.argmax(outputs.logits, dim=-1)

# Decode predictions
label_dict = {0: "neutral", 1: "positive", 2: "negative"}
predicted_labels = [label_dict[pred.item()] for pred in predictions]

# Add predicted sentiments to the dataframe
df['predicted_sentiment'] = predicted_labels

# Save the predictions to a new CSV file
df.to_csv("predicted_sentiments.csv", index=False)

print("Predicted sentiments saved to predicted_sentiments.csv")
```

```python
import numpy as np
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification
import torch

# Load test dataset
test_df = pd.read_csv("/content/test_aiquest - englesh.csv")

# Preprocess text data
test_df['text'] = test_df['text'].str.lower()

# Load pre-trained BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('./sentiment_model')

# Tokenize text data
test_encodings = tokenizer(test_df['text'].tolist(), truncation=True, padding=True)

# Convert data to PyTorch format
class TestDataset(torch.utils.data.Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        return item

    def __len__(self):
        return len(self.encodings['input_ids'])

test_dataset = TestDataset(test_encodings)

# Load trained BERT model
model = BertForSequenceClassification.from_pretrained('./sentiment_model')

# Make predictions
model.eval()
predictions = []
with torch.no_grad():
    for batch in torch.utils.data.DataLoader(test_dataset, batch_size=8):
        outputs = model(**batch)
        logits = outputs.logits
        preds = torch.argmax(logits, dim=1).tolist()
        predictions.extend(preds)

# Decode predictions
label_encoder = LabelEncoder()
label_encoder.classes_ = np.array(['negative', 'neutral', 'positive'])  # Convert to NumPy array
decoded_predictions = label_encoder.inverse_transform(np.array(predictions)) # Convert predictions to numpy array
# Save predictions to submission.csv
submission_df = pd.DataFrame({'id': test_df['id'], 'sentiment': decoded_predictions})
submission_df.to_csv('submission.csv', index=False)
```
